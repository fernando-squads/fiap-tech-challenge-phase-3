**Visão Geral**

Após o sucesso na automação de análises de exames e textos clínicos,
o hospital quer avançar para um nível superior de personalização: criar um
**assistente virtual médico, treinado com os dados próprios do hospital**,
capaz de auxiliar nas condutas clínicas, responder dúvidas de médicos e
sugerir procedimentos com base nos protocolos internos.

Além disso, a ideia é organizar fluxos de decisão automatizados e
seguros, onde, por exemplo, ao receber informações sobre um paciente, o
sistema possa acionar diferentes etapas, como verificar exames pendentes,
sugerir tratamentos e emitir alertas para a equipe médica — tudo isso
coordenado com **LangChain**.

**Objetivos do Projeto**



1.   Fine-tuning de LLM com dados médicos internos
2.   Criação de assistente médico com LangChain
3.   Segurança e validação
4.   Organização do código





**Dataset**

Foram utilizados dois datasets para a implementação do desafio: **PubMedQA** e **MedQuAD**. A partir deles, foi criado um dataset unificado contendo os atributos instruction e output.

Para o dataset **PubMedQA**, adotou-se a seguinte estratégia:

*   O atributo instruction foi gerado pela concatenação dos campos QUESTION e CONTEXTS;
*   O atributo output foi definido a partir do campo LONG_ANSWER.


Já para o dataset **MedQuAD**, foi necessário inicialmente converter os dados para o formato JSON. Em seguida, aplicou-se a seguinte estrutura:

*   O atributo instruction foi baseado no campo Question;
*   O atributo output foi derivado do campo Answer.

A utilização combinada desses dois datasets teve como objetivo a construção de uma base de dados mais robusta para o treinamento do modelo. Considerando a abrangência e qualidade do dataset unificado, não houve necessidade de incorporar dados adicionais, como protocolos médicos institucionais, modelos de laudos, receitas ou procedimentos internos hospitalares.
